## 1.1 Create a two-dimensional vector and plot it on a graph

```
#import libraries for assignment
import math
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

```
#setting the array that will be the end point of the arrow
shot=np.array([.5,.9])
#ploting the arrow starting point (0,0)next 2 point are derived from the first 
#and second position of list items in the array then followed by additional
#modifiers to adjust the look of the arrow
plt.arrow(0,0,shot[0],shot[1],overhang=.3,head_width=.02, head_length=.06, color='#0f6ab9',length_includes_head= True,shape='left')
plt.show()
```

## 1.2 Create a three-dimensional vecor and plot it on a graph

```
#setting up an array with an intial 3d plane starting value of (0,0,0) and a derectional value of my chosing equaling 6 list items
longshot = np.array([0,0,0,.9,.5,.7])
#setting the longshot array values to understandable values for the ploting with xyz starting values which would be the (0,0,0) starting point and directions in the x, y and z directions out from that point
X,Y,Z,U,V,W=zip(longshot)
#changing the background colors to black and text to white
plt.style.use('dark_background')
#setting up the graph paramiter
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
#using the predetermined values to place the arrow onto the now generated 3d space
ax.quiver(X, Y, Z, U, V, W, length=1, color = 'crimson')
# Setting the bounds of the graph space
ax.set_xlim([0, 2])
ax.set_ylim([0, 2])
ax.set_zlim([0, 2])
#lableing the 3 axes of the graph
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```

## 1.8 Find $||g||$ and then find $||h||$. Which is longer?

\begin{align}
\vec{g} = \begin{bmatrix} 1 \\ 1 \\ 1 \\ 8 \end{bmatrix}
\qquad
\vec{h} = \begin{bmatrix} 3 \\3 \\ 3 \\ 3 \end{bmatrix}
\end{align}

g is longer

```
g=np.array([[1],
            [1],
            [1],
            [8]])
h=np.array([[3],
            [3],
            [3],
            [3]])
normg=np.linalg.norm(g)
normh=np.linalg.norm(h)
print(normg)
print(normh)
```
